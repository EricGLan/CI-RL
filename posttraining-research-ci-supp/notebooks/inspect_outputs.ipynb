```python
import os
import json
import pandas as pd
from tempfile import TemporaryDirectory
# pip install git+https://github.com/microsoft/responsible-ai-toolbox-privacy.git#egg=privacy_estimates[pipelines]
from privacy_estimates.experiments.aml import JobList, Job
```

```python
job_url = "https://ml.azure.com/runs/e135af27-a6eb-41f3-8264-5c481059e08f?wsid=/subscriptions/acc09744-1ee3-4242-b375-93421c63af0c/resourcegroups/Singularity/workspaces/M365Research&tid=72f988bf-86f1-41af-91ab-2d7cd011db47"
```

```python
job = Job.from_url(job_url)
```

```python
def get_results(pipeline: Job) -> dict:
    results = {}
    privacy_lens_final_action = pipeline.get_node("privacylens_final_action")
    with TemporaryDirectory() as tempdir:
        privacy_lens_final_action.download_output(name="results", path=tempdir)
        print(f"Output: {os.listdir(tempdir)}")
        results["privacy_enhanced"] = pd.read_csv(
            os.path.join(tempdir, "privacy_enhanced.csv")
        )
        # seems to have the wrong extension
        results["privacy_enhanced_judge_leakage"] = pd.read_json(
            os.path.join(tempdir, "privacy_enhanced_judge_leakage.csv")
        )
        results["privacy_enhanced_full_responses"] = pd.read_json(
            os.path.join(tempdir, "privacy_enhanced_full.json")
        )

    # Do something else with the pipeline and add it to results
    return results
results = get_results(job)
```
